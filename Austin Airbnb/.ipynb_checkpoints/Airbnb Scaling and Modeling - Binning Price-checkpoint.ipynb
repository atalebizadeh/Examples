```python
data = data[data['price'] < data['price'].quantile(.85)]
```

```python
features_list = [
    'property_type',
    'room_type',
    'bathrooms',
    'bedrooms',
    'host_is_superhost',
    'accommodates',
    'number_of_reviews',
    'extra_people',
]
features = data[features_list]
```

```python
features = pd.get_dummies(features, drop_first=True)
features.head()
```

```python
scaler = StandardScaler()
numerical = [
    'bathrooms',
    'bedrooms',
    'accommodates'
]
X_scaled = features.copy()
X_scaled[numerical] = scaler.fit_transform(X_scaled[numerical])
```

```python
y = data['price'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=.3, random_state=42)
```

```python
print(X_train.shape, X_test.shape)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
```

```python
lr.intercept_
```

```python
lr.get_params
```

```python
lr.coef_
```

```python
y_test = y_test
resid = y_test - y_pred
resid_prop = resid / y_test
plt.scatter(np.arange(len(resid_prop)), resid_prop, s=15)
plt.axhline(0)
plt.title('Residual proportions (resid / actual Weight)')
plt.xlabel('Index of row in data')
plt.ylabel('Error proportion');
```

```python
r2_score(y_test, y_pred)
```
