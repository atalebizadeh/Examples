```python
import pandas as pd
import pandasql as ps
```

```python
# cd which is short for CovidDeaths
cd = pd.read_excel('data/CovidDeaths.xlsx')
cd.head()
```

```python
# cv which is short for CovidVaccinations
cv = pd.read_excel('data/CovidVaccinations.xlsx')
cv.head()
```

```python
#to read cd table
ps.sqldf("select * from cd where continent is not null order by 3,4")
```

```python
#to read cv table
ps.sqldf("select * from cv where continent is not null order by 3,4")
```

```python
ps.sqldf("select location, date, total_cases, new_cases, total_deaths, population from cd where continent is not null order by 1,2")
```

```python
# shows likelihood of dying if you contract covid in US
ps.sqldf("select location, date, total_cases, total_deaths, (total_deaths/total_cases)*100 as DeathPercentage from cd where location like '%states%' and continent is not null  order by 1,2")
```

```python
# total cases vs the population
ps.sqldf("select location, date, population, total_cases, (total_cases/population)*100 as PercentPopulationInfected from cd where location like '%states%' and continent is not null order by 1,2")
```

```python
# looking for countries with highest infection rate compated to population
ps.sqldf("select location, population, MAX(total_cases) as HighestInfectionCount, MAX((total_cases/population))*100 as PercentPopulationInfected from cd where continent is not null group by location, population order by PercentPopulationInfected desc")
```

```python
# showing countries with highest death count per population
ps.sqldf("select location, MAX(cast(total_deaths as int)) as TotalDeathCount from cd where continent is not null group by location order by TotalDeathCount desc")
```

```python
# showing continent with highest death count
ps.sqldf("select continent, MAX(cast(total_deaths as int)) as TotalDeathCount from cd  where continent is not null group by continent order by TotalDeathCount desc")
```

```python
# let's break things down by continent
ps.sqldf("select location, MAX(cast(total_deaths as int)) as TotalDeathCount from cd  where continent is null group by location order by TotalDeathCount desc")
```

```python
# Global numbers
ps.sqldf("select location, date, total_cases, total_deaths, (total_deaths/total_cases)*100 as DeathPercentage from cd where continent is not null order by 1,2")
```

```python
ps.sqldf("select date, SUM(new_cases) as TotalCases, SUM(cast(new_deaths as int)) as TotalDeaths, SUM(cast(new_deaths as int))/SUM(new_cases)*100 as DeathPercentage from cd where continent is not null group by date order by 1,2")
```

```python
ps.sqldf("select SUM(new_cases) as TotalCases, SUM(cast(new_deaths as int)) as TotalDeaths, SUM(cast(new_deaths as int))/SUM(new_cases)*100 as DeathPercentage from cd where continent is not null order by 1,2")
```

```python
# let's look at cv - Covid Vaccinations table
ps.sqldf("select * from cv")
```

```python
# let's join two tables
ps.sqldf("select * from cd join cv on cd.location = cv.location and cd.date = cv.date")
```

```python
# looking at total population vs vaccinations
ps.sqldf("select cd.continent, cd.location, cd.date, cd.population, cv.new_vaccinations, SUM(cast(cv.new_vaccinations as int)) OVER(partition by cd.location order by cd.location, cd.date) as RollingPeopleVaxed from cd join cv on cd.location = cv.location and cd.date = cv.date where cd.continent is not null order by 2,3")
```

```python
# using CTEs
ps.sqldf("WITH PopvsVac (continent, location, date, population, new_vaccinations, RollingPeopleVaxed)"
         "AS (select cd.continent, cd.location, cd.date, cd.population, cv.new_vaccinations, SUM(cast(cv.new_vaccinations as int)) OVER(partition by cd.location order by cd.location, cd.date) as RollingPeopleVaxed from cd join cv on cd.location = cv.location and cd.date = cv.date where cd.continent is not null)"
"SELECT *, (RollingPeopleVaxed/population)*100 PopulationVaxedPercentage FROM PopvsVac")
```
